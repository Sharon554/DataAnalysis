# Groundwater is a vital resource for agricultural, industrial, and domestic purposes, especially in regions like Tamil Nadu, where a significant portion of irrigation relies on tubewells. This dataset provides comprehensive information on groundwater and tubewell schemes implemented across various districts in Tamil Nadu, India. It offers insights into the location, ownership, financial details, operational status, irrigation potential, and groundwater recharge contributions of these schemes
---------------------------------------------------------------------------------------------------------

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv("MI6_GW_STW_TN.csv", low_memory=False)
```

```python
df.fillna({
    'scheme_owner_name': 'Unknown',  
    'tubewell_type': 'Unknown',  
    'command_area_name': 'Not Specified',  
}, inplace=True)
```

```python
df.replace({'': 'Unknown'}, inplace=True)
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(0)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df[numeric_cols] = df[numeric_cols].astype(float)
```

```python
print(df.head())
```

```python
df.info()
```

```python
df = df.fillna(0)
```

```python
df = df.infer_objects(copy=False)
```

```python
print(df.isnull().sum())
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.figure(figsize=(12, 6))
df['district_name'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Districts with Most Tube Wells')
plt.xlabel('District Name')
plt.ylabel('Number of Tube Wells')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['tubewell_type'].value_counts().plot.pie(autopct='%1.1f%%', colors=['blue', 'orange', 'green'])
plt.title('Distribution of Tubewell Types')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['total_owners_holding'])
plt.title('Distribution of Total Owners Holding')
plt.ylabel('Total Owners Holding')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['s_no_of_scheme'], df['total_owners_holding'], marker='o', linestyle='-')
plt.title('Trend of Schemes Over Serial Numbers')
plt.xlabel('Scheme Serial Number')
plt.ylabel('Total Owners Holding')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x=df['scheme_type_name'], y=df['total_owners_holding'])
plt.title('Violin Plot of Scheme Type vs Owners Holding')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['total_owners_holding'], df['ipu_total'], alpha=0.5, color='purple')
plt.title('Owners Holding vs IPU Total')
plt.xlabel('Total Owners Holding')
plt.ylabel('IPU Total')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['total_owners_holding'], bins=30, color='green', edgecolor='black')
plt.title('Histogram of Owners Holding')
plt.xlabel('Total Owners Holding')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df['village_name'].value_counts().head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Villages with Most Schemes')
plt.xlabel('Village Name')
plt.ylabel('Number of Schemes')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df['scheme_owner_name'].value_counts().plot(kind='bar', color='coral')
plt.title('Schemes Owned by Different Groups')
plt.xlabel('Owner Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='district_name')
plt.title('Distribution of Number of Schemes by District', fontsize=16)
plt.xticks(rotation=45)
plt.xlabel('District Name')
plt.ylabel('Number of Schemes')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='scheme_type_name', y='total_owners_holding')
plt.title('Distribution of Total Owners Holding by Scheme Type', fontsize=16)
plt.xlabel('Scheme Type')
plt.ylabel('Total Owners Holding')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='tubewell_type')
plt.title('Distribution of Tubewell Types', fontsize=16)
plt.xlabel('Tubewell Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='scheme_owner_name')
plt.title('Number of Schemes by Scheme Owner Name', fontsize=16)
plt.xlabel('Scheme Owner Name')
plt.ylabel('Number of Schemes')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(data=df, x='district_name')
plt.title('Distribution of Number of Schemes by District', fontsize=16)
plt.xticks(rotation=45)
plt.xlabel('District Name')
plt.ylabel('Number of Schemes')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='scheme_type_name')
plt.title('Number of Schemes by Scheme Type', fontsize=16)
plt.xlabel('Scheme Type')
plt.ylabel('Number of Schemes')
plt.xticks(rotation=45)
plt.show()
```

```python
tubewell_counts = df['tubewell_type'].value_counts()
plt.figure(figsize=(8, 8))
tubewell_counts.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('Set3', len(tubewell_counts)))
plt.title('Distribution of Tubewell Types', fontsize=16)
plt.ylabel('')  # To hide the y-label
plt.show()
```
